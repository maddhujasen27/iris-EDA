### IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### LOADING DATA

```python
from sklearn.datasets import load_iris 
iris = load_iris()
```

<img src="https://miro.medium.com/max/1000/1*Hh53mOF4Xy4eORjLilKOwA.png" width = '300'>

```python
df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
df
```

```python
df['species'] = iris.target
```

#### SCATTER PLOT

```python
plt.scatter(df['petal length (cm)'],df['sepal length (cm)'], color = 'orange')
plt.xlabel("petal length (cm)")
plt.ylabel("sepal length (cm)")
plt.title("Plotting petal length (cm) VS sepal length (cm) in Scatter Plot")
plt.show()
```

```python
plt.plot(df['petal width (cm)'],df['sepal width (cm)'])
plt.scatter(df['petal width (cm)'],df['sepal width (cm)'], color = "red")
plt.xlabel("petal width (cm)")
plt.ylabel("sepal width (cm)")
plt.title("Petal width (cm) VS Sepal width (cm)")
plt.show()
```

```python
plt.scatter(df['petal width (cm)'],df['petal length (cm)'], color = "red")
plt.xlabel("petal width (cm)")
plt.ylabel("petal length (cm)")
plt.title("Petal width (cm) VS Petal length (cm)")
plt.show()
```

```python
plt.scatter(df['sepal width (cm)'],df['sepal length (cm)'], color = "green")
plt.xlabel("sepal width (cm)")
plt.ylabel("sepal length (cm)")
plt.title("Sepal width (cm) VS Sepal length (cm)")
plt.show()
```

```python
plt.figure(figsize = (9, 7))
plt.scatter(df['sepal length (cm)'], df['petal length (cm)'])
plt.scatter(df['sepal width (cm)'], df['petal length (cm)'])
plt.legend(['length','width'])
```

```python
plt.figure(figsize = (9, 7))
sns.swarmplot(x = 'species', y = 'petal length (cm)', data = df)
```
